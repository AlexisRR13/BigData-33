# **0.1 - Construccion de la muestra M**

La población de estudio utilizada en este proyecto corresponde al conjunto de datos AI vs Human Text, obtenido desde Kaggle. En etapas previas del proyecto se realizó un análisis exploratorio de datos con el fin de evaluar su calidad, verificando la existencia de valores faltantes, registros duplicados, inconsistencias en los tipos de datos y posibles anomalías que pudieran afectar el proceso de modelado. Como resultado de dicho análisis se concluyó que el conjunto de datos presenta un nivel adecuado de calidad y no requiere procesos adicionales de limpieza o depuración para los objetivos de esta investigación.

Con la finalidad de enriquecer el análisis y generar criterios de particionamiento que permitan construir muestras representativas, se incorporó una variable derivada denominada longitud del texto, calculada a partir del número de palabras contenidas en cada documento. Posteriormente, mediante el uso de cuantiles, los textos fueron clasificados en tres categorías: corto, mediano y largo. Esta variable complementa la variable objetivo (generated), permitiendo caracterizar la población de manera más detallada y construir estratos que preserven la diversidad existente en los datos. Se recalca que el muestreo en estratos fue el más apropiado para nuestra base de datos, ya que un muestreo aleatorio hubiera generado patrones que pueden sobre representar o sub representar ciertos comportamientos de los datos. Muestro estratificado garantiza que cada sub grupo sea representado apropiadamente.

En las etapas anteriores se trabajó con una muestra reducida de la población con el propósito de disminuir los tiempos de procesamiento durante las pruebas iniciales. Sin embargo, para esta etapa se decidió utilizar una muestra equivalente al 60% de la población disponible. Esta decisión se tomó con el objetivo de maximizar la representatividad de los resultados obtenidos y evitar la pérdida de información potencialmente relevante para el entrenamiento y evaluación de los modelos de aprendizaje automático.
Ya que se realizaron pruebas de tiempo de ejecución con la totalidad de la población y se compararon contra la evaluación de los modelos, sin tener gran variación en el resultado.

El uso de la población completa implica un incremento considerable en los requerimientos computacionales y en los tiempos de ejecución. Por este motivo, se implementaron diversas estrategias de optimización dentro del entorno PySpark, incluyendo el almacenamiento en memoria mediante técnicas de cache/persist, la reducción de operaciones redundantes y el aprovechamiento del procesamiento distribuido proporcionado por Spark. Estas medidas permiten trabajar con grandes volúmenes de datos de manera eficiente sin comprometer la representatividad de la muestra utilizada.

Finalmente, para garantizar que las distribuciones originales de la población se conservaran durante las etapas posteriores del proyecto, se construyeron estratos a partir de las variables generated y longitud, los cuales serían utilizados posteriormente para la generación de los conjuntos de entrenamiento y prueba mediante muestreo estratificado.

```python
#1.1 Se inicia la sesión en spark y se configura de esta manera:
#.master("local[*]") -> se corre spark en esta computadora localmente y se usan todos los GPU disponibles.
#.appName("Proyecto_Etapa3") -> etiqueta de sesión.
#.config("spark.driver.memory", "12g") - > se indica la cantidad de RAM que spark tiene disponible de usar
#.config("spark.sql.shuffle.partitions", "4") -> Se condiguró con 4 para reducir la sobrecarga por el número excesivo de particiones que Spakr crea por defecto. Ya que en este momento se está corriendo en una computadora local y no un cluster.
#spark.conf.set("spark.sql.repl.eagerEval.enabled", True) -> se configura mejorar la visualización de las tablas
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .master("local[*]")
    .appName("Proyecto_Etapa3")
    .config("spark.driver.memory", "12g")
    .config("spark.sql.shuffle.partitions", "4")
    .getOrCreate()
)

spark.conf.set("spark.sql.repl.eagerEval.enabled", True)
```

```python
#1.2 Descarga del dataset desde Kaggle.
#Se puede accesar directamente datasets de kaggle con la libreria "kagglehub" y su función dataset_download.
import kagglehub
path = kagglehub.dataset_download("shanegerami/ai-vs-human-text")
print("Path to dataset files:", path)
```

```python
#1.3 - Se crea el dataframe del dataset:
#Se utilizó spark.read.csv que permite extraer el dataset desde un arhcivo .csv.
#En la función spark.read.csv se especifica como está configurado el archivo.
#De una exploración inicial se conoce que: el archivo tiene encabezados que se buscan mantener.
#inferSchema indica que se va a inferir el tipo de dato de acuerdo al contenido.
#multiLine indica que se espera que los textos del .csv tengan multilínea y que no haga una nueva fila.
#quote "" permite mantener en una sola fila strings que tengan commas. el requisito es que el string empiece y termine con "".
#escape "" permite lidiar con comillas dentro de un texto, para que no lo convierta en otra fila.
#Se utilizó df.persist para que el dataframe pueda guardarse en RAM y disco duro y que no haga crash si se llega a llenar el RAM.
df = spark.read.csv(
    path,
    header=True,
    inferSchema=True,
    multiLine=True,
    quote='"',
    escape='"'
)

from pyspark import StorageLevel

df = df.persist(StorageLevel.MEMORY_AND_DISK)

# Force cache
df.count()
```

```python
#1.4 - Se crea columna "num_palabras"
#Dado que el dataset originalmente sólo contiene las columnas "text" y "generated" necesitamos crear más variables del que pueda entrenar el algoritmo.
#Se optó por el número de palabras, ya que se encontró de un análisis preliminar de que la IA tiende a ser más conciso que los textos generador por humanos.
#Se utilizó la función split para dividir textos con el delimintador " " (espacio) y guardarlos en una lista.
#Se utilizó la función size para medir la cantidad de elementos de una lista.
from pyspark.sql.functions import split, size

df = df.withColumn(
    "num_palabras",
    size(split(df["text"], " "))
)
```

```python
#1.5 - Se crean bins con la variables num_palabras para categorizar comentarios en: corto, mediano y largo.
#Con el objetivo de estratificar la población para mantener una misma proporción de comentarios en la muestra y posteriormente en los conjuntos test y train.
#Se optó hacerlo por terciles, donde q1 sería el 33% y q2 el 66%.
#Se clasifica entonces que los comentarios serán cortos si tienen menos de 304 palabras.
#Los comentairos serán medianos si están entre 304 y 415 palabras.
#Los comentarios serán largos si son mayores a 415 palabras.
#Esta categoría se guarda en una nueva columna llamada "longitud"
quantiles = df.approxQuantile(
    "num_palabras",
    [0.33, 0.66],
    0.01
)

q1, q2 = quantiles


from pyspark.sql.functions import when

df = df.withColumn(
    "longitud",
    when(df.num_palabras < q1, "corto")
    .when(df.num_palabras < q2, "mediano")
    .otherwise("largo")
)

print(q1, q2)
```

```python
#1.6 - Creación de estratos:
#Para esta parte, se crea una nueva columna llamada "estratos".
#la columna "estratos" será la combinación de [0 o 1] de generated y [corto, mediano o largo] de "longitud".
#finalmente, se devuelve los valores sin duplicados de "estrato" para darnos cuenta de todas las posibles combinaciones.
#En este caso, las combinaciones son principalmente 6: 0.0_largo, 1.0_corto, 1.0_largo, 1.0_mediano, 0.0_mediano, 0.0_corto.
#Es importante conocer todos los estratos, ya que se buscará que la muestra y particiones train - test, tengan la misma distribución que la población.
from pyspark.sql.functions import concat_ws, col

df = df.withColumn(
    "estrato",
    concat_ws("_", col("generated"), col("longitud"))
)

df.select("estrato").distinct().show(truncate=False)
```

```python
#1.7 - En este paso se determina el porcentaje de la población que se utilizará para la muestra.
#Como se indicó en el inicio de esta sección y en las instrucciones de la actividad, se optará por usar el 60% de la población.
#Finalmente, se confirma que la muestra es de un 60% de la poblacion.
#Se recalca en este punto que se utilizó
sample_fraction = 0.6

muestra = df.sampleBy( #muestreo estratificado
    "estrato", #columna que define los grupos (estratos)
    fractions={
        "0.0_corto": sample_fraction,
        "0.0_mediano": sample_fraction,
        "0.0_largo": sample_fraction,
        "1.0_corto": sample_fraction,
        "1.0_mediano": sample_fraction,
        "1.0_largo": sample_fraction
    },
    seed=42 #semilla para asegurar reproducibilidad del muestreo
)


muestra.count()
```

#**0.2 - Construccion Train – Test**
* Justificar partición 20%-80%:

Si bien el porcentaje de test es un parámetro que también se debe de ajustar para mejorar los resultados,se notó que un 20% fue suficiente para aún así mantener una proporción del estrato menos ocurrente (1.0_largo) igual a la población. Esto se comprueba más adelante que mostramos la cantidad de elementos del estrato 1.0_largo en test (7230) que representa 7.4% del total. Comparado con el 7.31% de la población, nos  parece que mantenemos una proporción deaseada. Por eso, 20% es suficiente y se sacará provecho del 80% restante para entrenar el modelo.

```python
#2.1 - En este bloque de código se elige el porcentaje train-test. En este caso, test será 20% y train 80% de la muestra.
#para esto se usó la función "sampleBy" es lo que nos permite hacer la partición estratificada y no random.
#funciona indicando a spark que se tome un 20% de cada subgrupo para formar el test.
#el train se crea con los datos restantes que no fueron seleccionados para el test.
#finalmente, hacemos otra vez uso de la función creada "estrato_percentages" para verificar que tanto train y test tienen proporciones semejantes.
#Como se mencionó anteriormente, el estrato que más se debe de cuidar de no sub representar es 1.0_largo.
#Se comprueba que con este muestreo estratificado se mantiene una buena cantidad de datos en este estrato 1.0_largo.
#Por lo que no hay ninguna razón para aumentar el test a otro porcentaje y mejor se optará de usar la mayor cantidad de datos al train para aprender mejor de los patrones de los datos.

test_fraction = 0.2 #Fraccion para el conjunto de prueba de 20%

#Obtencion de los valores distintos de la columna "estrato" en la muestra
strata = (
    muestra
    .select("estrato")
    .distinct()
    .collect()
)
#Diccionario con la fraccion deseada para cada estrato
fractions = {
    row["estrato"]: test_fraction
    for row in strata
}

#Conjunto de prueba (test_data) usando muestreo estratificado
test_data = muestra.sampleBy(
    "estrato",
    fractions=fractions,
    seed=42 # Semilla para reproducibilidad
)
#Conjunto de entrenamiento (muestra original- datos de prueba)
train_data = muestra.subtract(test_data)

```

```python
#2.2 - Dado que estamos usando 60% de los datos de un dataset grande, se guardarán los datasets y se creará una variable con su conteo.
#Simplemente para no tener que repetir este proceso cuando se vuelva a necesitar esos datos.
#También, se comprueba el tamaño esperado para el conjunto train y test.

#Guarda en memoria el DataFrame
muestra.cache()
train_data.cache()
test_data.cache()

#Conteo de filas de cada dataset y se almacenan en variables
muestra_count = muestra.count()
train_count = train_data.count()
test_count = test_data.count()

#Impresion de conteos para verificar tamaños esperados
print("Muestra:", muestra_count)
print("Train :", train_count)
print("Test  :", test_count)
```

#**0.4-Entrenamiento de Modelos de Aprendizaje**

Para este problema se optó por trabajar con algortimos supervisados. La principal razón es que nuestros datos ya contienen un label que nos indican si el texto es generado por humanos o por IA. En actividades anteriores, se intentaron modelos no supervisados y los resultados fueron considerablemente peores a los supervisados. Con esto se afirma que para usar modelos no supervisados se requiere de métodos más avanzados y de un dataset de mayor tamaño.

El procedimiento de entrenamiento es el siguiente:
* Se divide los comentarios en palabras (tokenizer)
* Las palabras se convierten en vectores (HashingTF)
* Se forma la matriz IDF (Inverse Document Frequency)
* Se utiliza Regressión Logística ya que es un modelo computacionalmente eficiente. (especialmente útil en este problema donde la cantidad de experimentos que se puede hacer es límitado).
* Se crea un pipeline con esta secuencia de pasos (tokenizer, hashingTF, IDF y Lr). Para estandarizar el proceso y evitar data leakage.
* Finalmente, para encontrar los mejores hiperparámetros. Se hace un cross validation. Sin embargo, dado que múltiples búsquedas pueden hacer muy largo el entrenamiento, se limitó a sólo buscar entre [0.01, 1] para la fuerza de regulación (regParam) y [0.0, 0.5] para controlar el tipo de regulación 0.0 -> Ridge L2 y 0.5 -> Combinación entre Ridge y Lasso.
* Se elige la métrica principal, que como se mencionó en el apartado anterior, para este problema será F1.
* Se hace el cross validation donde se elige como fold:3, es decir, que se divide en 3 partes el training data, 2 partes para entrenar y 1 para validar. esto se va a iterar con cada uno de los hiperparámetros establecidos en el punto anterior.
* Por lo tanto, el modelo será entrenado 12 veces. Lo que se incrementa bastante el tiempo de ejecución. se agregó un apartado para estimar el tiempo que tardó como referencia.


#**1.- Justificacion estadística del valor de (K)**

El valor de K impacta directamente el sesgo y la varianza del modelo:

- K pequeño → mayor sesgo, menor varianza
- K grande → menor sesgo, mayor varianza y alto costo computacional

En este caso:

- Tamaño del dataset: grande (Big Data, 60% poblacion total)
- Muestreo: estratificado - reduce varianza
- Algoritmo: regresión logistica (baja varianza)

Por lo tanto, no es necesario un fold alto para estabilizar resultados.

### Validacion numerica del tamaño de los folds

Sea N el tamaño del entrenamiento:

Cada fold tiene tamaño ≈ N / k

Si k = 3:
- Cada fold ≈ 33% de los datos
- Suficientemente grande para ser representativo

Si k = 10:
- Cada fold ≈ 10%
- Riesgo de perdida de representatividad en estratos minoritarios

### Validacion con estratos

Debido al muestreo estratificado en variables:
- generated
- longitud

Cada fold mantiene aproximadamente la misma distribucion:

P(estrato_i en fold_j) ≈ P(estrato_i en poblacion)

Esto garantiza representatividad estadistica.

Por lo que podemos concluir que:

Se selecciona **k = 3**, ya que:
- Maximiza eficiencia computacional
- Mantiene estabilidad en resultados
- Garantiza representatividad por estratos

#**2.- Construccion de los k-folds**

```python
from pyspark.ml.feature import (
    Tokenizer,
    HashingTF,
    IDF
)

from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline

from pyspark.ml.tuning import (
    ParamGridBuilder,
    CrossValidator
)

from pyspark.ml.evaluation import (
    MulticlassClassificationEvaluator
)

#Procesamiento del texto
tokenizer = Tokenizer(
    inputCol="text",         #Columna de entrada con el texto
    outputCol="words"        #Columna de salida con las palabras tokenizadas
)

hashingTF = HashingTF(
    inputCol="words",         #Uso de palabras Tokenizadas
    outputCol="rawFeatures",  #Vector de frecuiencias
    numFeatures=2000          #Numero de caracteristicas (dimension del espacio)
)

idf = IDF(
    inputCol="rawFeatures",   #Entrada: frecuencias
    outputCol="features"      #Salida: TF - IDF (ponderacion de importancia)
)

#Modelo Regresion logistica
lr = LogisticRegression(
    featuresCol="features",    #Caracteristica de entrada
    labelCol="generated",      #Columna con las etiquetas
    maxIter=20                 #Numero maximo de iteraciones
)

#Pipeline
pipeline = Pipeline(
    stages=[
        tokenizer,             #Tokenizacion
        hashingTF,             #Vectorizacion
        idf,                   #Ponderacion TF-IDF
        lr                     #Clasificacion con modelo de regresion logistica
    ]
)

#Definicion de parametros para Cross validation
paramGrid = (
    ParamGridBuilder()
    .addGrid(lr.regParam, [0.01, 0.1])        #Parametro de regularizacion
    .addGrid(lr.elasticNetParam, [0.0, 0.5])  #Mezcla L1/L2
    .build()
)

#Evaluacion metrica de desempeño
evaluator = MulticlassClassificationEvaluator(
    labelCol="generated",                      #Etiquetas reales
    predictionCol="prediction",                #Predicciones del modelo
    metricName="f1"                            #Metrica F1 (balannce entre precision y recall)
)
k = 3
#Cross validation
cv = CrossValidator(
    estimator=pipeline,                        #Pipeline completo
    estimatorParamMaps=paramGrid,              #parametros
    evaluator=evaluator,                       #Metrica de evaluacion
    numFolds=k,                                #Numero de folds K
    seed=42                                    #Semilla para reproducibilidad
)

#Entrenamiento del modelo con medicion de tiempo
import time

start = time.time()

cvModel = cv.fit(train_data)   #Entrena el modelo con los datos de entrenamiento

training_time = time.time() - start

print(f"Training time: {training_time:.2f} seconds")


#Obtencion del mejor modelo y sus parametros
bestModel = cvModel.bestModel

bestLR = bestModel.stages[-1]   #Ultima etapa del pipeline (Regresion logistica)

print("\nBest Parameters Found")
print("---------------------")
print("regParam       :", bestLR.getRegParam())
print("elasticNetParam:", bestLR.getElasticNetParam())
#prediccion sobre el conjunto de prueba
predictions = cvModel.transform(test_data)
```

#**3.- Funcionamiento interno de CrossValidator en PySpark**

El método CrossValidator implementa el siguiente flujo:

1. Divide el dataset en k particiones aleatorias
2. Para cada combinacion de hiperparametros:
   - Entrena con k-1 folds
   - Valida con el fold restante
3. Promedia la metrica (F1)
4. Selecciona el mejor modelo

### Consideracion para Big Data

Para evitar:
- re-calculo innecesario
- sobrecarga de memoria

Se aplico:
- cache() sobre train_data
- reducción de particiones en Spark

### Evaluacion por fold

Durante la validación cruzada:

- Cada modelo es evaluado en k folds
- Se obtiene un F1 por cada iteración
- PySpark retorna el promedio de estos valores

Esto permite:

- Medir estabilidad del modelo
- Detectar variabilidad

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator, BinaryClassificationEvaluator

# F1
f1 = MulticlassClassificationEvaluator(
    labelCol="generated",
    predictionCol="prediction",
    metricName="f1"
).evaluate(predictions)

# Accuracy
accuracy = MulticlassClassificationEvaluator(
    labelCol="generated",
    predictionCol="prediction",
    metricName="accuracy"
).evaluate(predictions)

# AUC
auc = BinaryClassificationEvaluator(
    labelCol="generated",
    rawPredictionCol="rawPrediction",
    metricName="areaUnderROC"
).evaluate(predictions)

print("F1:", f1)
print("Accuracy:", accuracy)
print("AUC:", auc)
```
